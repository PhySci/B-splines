```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline as Spline
%matplotlib inline
```

```python
# read data
data = np.genfromtxt('data/data.csv', delimiter=',')
xData = data[:,0]
yData = data[:,1]
weights = data[:,2]
sl = data[:,3]
```

```python
data.shape
```

```python
# smooth parameter
p = 0.003355272021954153
```

```python
plt.figure(figsize=[15,5])
plt.plot(xData,yData,'ro')
plt.plot(xData,sl)
plt.xlabel('x')
plt.ylabel('y')
```

```python
fit = Spline(xData,yData, s = 10000)
pr = fit(xData)
plt.plot(xData,pr)
```

```python
plt.figure(figsize = [15,10])
plt.subplot(211)
plt.plot(xData,pr)
plt.plot(xData,yData)

plt.subplot(212)
plt.plot(xData,pr-yData)
```

```python
plt.plot(xData,yData-pr)
```

```python
fit = Spline(xData,yData, s = 10000000, w = weights)
pr = fit(xData)
plt.figure(figsize=[20,10])
plt.subplot(211)
plt.plot(xData,sl) 
plt.plot(xData,pr)
plt.legend(['MatLab','Scipy'])

plt.subplot(212)
plt.plot(xData,sl-pr)
plt.legend('Residual')
```
